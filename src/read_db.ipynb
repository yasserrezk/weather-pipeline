## 1 — Imports & connection

```python
import os
import sys
import pandas as pd
import matplotlib.pyplot as plt
import matplotlib.dates as mdates
from dotenv import load_dotenv
from sqlalchemy import create_engine, text

# Add project root to path so src.* imports work
sys.path.insert(0, os.path.abspath('..'))

load_dotenv('../.env')

DB_USER     = os.getenv('DB_USER')
DB_PASSWORD = os.getenv('DB_PASSWORD')
DB_HOST     = 'localhost'  # localhost when running outside Docker
DB_PORT     = os.getenv('DB_PORT', '5432')
DB_NAME     = os.getenv('DB_NAME')

engine = create_engine(
    f'postgresql://{DB_USER}:{DB_PASSWORD}@{DB_HOST}:{DB_PORT}/{DB_NAME}'
)

# Verify connection
with engine.connect() as conn:
    result = conn.execute(text('SELECT NOW()')).scalar()
    print(f'Connected to {DB_HOST}:{DB_PORT}/{DB_NAME}')
    print(f'DB time: {result}')
```

## 2 — Row count

```python
with engine.connect() as conn:
    count = conn.execute(text("SELECT COUNT(*) FROM weather_forecasts")).scalar()

print(f"Total rows in weather_data: {count:,}")
```

## 3 — View all rows (latest 100)

```python
with engine.connect() as conn:
    df = pd.read_sql(
        text("SELECT * FROM weather_forecasts ORDER BY forecast_time LIMIT 100"), conn
    )

print(f"{len(df)} rows returned")
df
```
